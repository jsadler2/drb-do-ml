```python
import pandas as pd
```

```python
def read_and_plot_training_curve(train_log_file, title=None):
    df = pd.read_csv(train_log_file)
    ax = df.set_index('epoch').plot(title=title)
    print(df.tail())
    print(df.min())
    return ax
```

```python
lstm_metab = "1_lstm_metab/nstates_10/nep_100/rep_0/train_log.csv"
lstm_metab_400 = "1_lstm_metab/nstates_10/nep_400/rep_0/train_log.csv"
tf_ops = "4_lstm_metab_tf_ops_pow/nstates_10/nep_100/rep_0/train_log.csv"
tf_ops_add = "4_lstm_metab_tf_ops_add/nstates_10/nep_100/rep_0/train_log.csv"
metab_multitask = "2_metab_multitask/nstates_10/nep_100/rep_0/train_log.csv"
```

```python
read_and_plot_training_curve(lstm_metab, "normal")
```

```python
read_and_plot_training_curve(lstm_metab_400, "normal, 400 epochs")
```

```python
read_and_plot_training_curve(tf_ops, "pow")
```

```python
read_and_plot_training_curve(metab_multitask, "metab multitask")
```

```python
read_and_plot_training_curve(tf_ops_add, "tf ops add")
```
